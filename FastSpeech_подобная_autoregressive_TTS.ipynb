```python
# –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫
import time
import soundfile as sf
from TTS.api import TTS
from IPython.display import Audio

# –¢–µ–∫—Å—Ç –¥–ª—è –æ–∑–≤—É—á–∫–∏
text = """
Data Science isn't just about algorithms‚Äîit's about explaining your results to both machines and people.
Fluent English helps you document code, present findings, and publish papers.
From Jupyter notebooks to international teams, language shapes your success.
Invest in English to unlock deeper collaboration, stronger research, and global impact.
"""

# –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏ (–∞–Ω–≥–ª–∏–π—Å–∫–∏–π –≥–æ–ª–æ—Å Tacotron2 DDC)
start_load = time.time()
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)
load_time = time.time() - start_load
print(f"‚úÖ –ú–æ–¥–µ–ª—å –∑–∞–≥—Ä—É–∂–µ–Ω–∞ –∑–∞ {load_time:.2f} —Å–µ–∫.")

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —Ä–µ—á–∏
start_gen = time.time()
wav = tts.tts(text)
gen_time = time.time() - start_gen
print(f"üéß –ê—É–¥–∏–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–æ –∑–∞ {gen_time:.2f} —Å–µ–∫.")

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ .wav
sf.write("output.wav", wav, 22050)
print("üìÅ –§–∞–π–ª —Å–æ—Ö—Ä–∞–Ω—ë–Ω –∫–∞–∫ output.wav")

# –í–æ—Å–ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ
Audio("output.wav")
```
