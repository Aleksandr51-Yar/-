```python
# Импорт библиотек
import time
import soundfile as sf
from TTS.api import TTS
from IPython.display import Audio

# Текст для озвучки
text = """
Data Science isn't just about algorithms—it's about explaining your results to both machines and people.
Fluent English helps you document code, present findings, and publish papers.
From Jupyter notebooks to international teams, language shapes your success.
Invest in English to unlock deeper collaboration, stronger research, and global impact.
"""

# Загрузка модели (английский голос Tacotron2 DDC)
start_load = time.time()
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)
load_time = time.time() - start_load
print(f"✅ Модель загружена за {load_time:.2f} сек.")

# Генерация речи
start_gen = time.time()
wav = tts.tts(text)
gen_time = time.time() - start_gen
print(f"🎧 Аудио сгенерировано за {gen_time:.2f} сек.")

# Сохранение в .wav
sf.write("output.wav", wav, 22050)
print("📁 Файл сохранён как output.wav")

# Воспроизведение
Audio("output.wav")
```
